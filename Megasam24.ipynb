```python
import requests
from bs4 import BeautifulSoup
import csv
```

```python
#Megasam24 - mleko - nazwa
page1=requests.get('https://megasam24.pl/pl/c/Mleko-swieze/257')
soup = BeautifulSoup(page1.text, 'html.parser')
marka_mleko_megasam24 = soup.findAll('span', attrs={'class':'productname'}) 
for m1 in marka_mleko_megasam24:
    print(m1.text.strip())
```

```python
#Megasam24 - mleko - cena
soup = BeautifulSoup(page1.text, 'html.parser')
cena_mleko_megasam24 = soup.findAll('div', attrs={'class':'price f-row'}) 
for c1 in cena_mleko_megasam24:
    print(c1.text.strip())
```
